### Cleaning Data

```python
#memperbaiki tipe data pada orders_df

#Untuk mengatasi masalah ini, kita akan mengganti tipe data ,Proses ini dapat dilakukan menggunakan function to_datetime() yang disediakan oleh library pandas. Berikut merupakan contoh kode untuk melakukannya.

datetime_columns = ["order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
#Kode di atas akan mengubah tipe data menjadi datetime. Untuk memastikan hal ini berjalan sesuai harapan, periksa kembali tipe data tersebut menggunakan method info().

orders_df.info()
```

```python
#ternyata terdapat missing value pada orders_df

# Untuk memastikannya, kita bisa menjalankan kode berikut.
orders_df.isna().sum()
```

```python
#secara umum terdapat tiga metode untuk mengatasi missing value yaitu dropping, imputation, dan interpolation.

orders_df[orders_df.order_approved_at.isna()]
```

```python
#Berdasarkan gambar di atas, dapat dilihat bahwa baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. Oleh karena itu, pada kasus ini, kita akan menggunakan metode imputation untuk menangani missing value.


orders_df.order_approved_at.value_counts()
```

```python
#Bagaimana performa penjualan dan revenue perusahaan dalam beberapa bulan terakhir?

#untuk menjawab pertanyaan ini kita dapat melakukan merge dari dataset orders_df dan order_payments_df berikut merupakan contohnya

company_revenue_df = pd.merge(
    orders_df, order_payments_df, on='order_id'
    )

company_revenue_df.head()
```

```python
# tambahkan month column agar mempermudah dalam membuat pivot table nya nanti
company_revenue_df['order_purchase_timestamp'] = pd.to_datetime(company_revenue_df['order_purchase_timestamp'])

company_revenue_df['month'] = company_revenue_df['order_purchase_timestamp'].dt.to_period('M')
company_revenue_df.head()

#nah setelah terlihat coloum month nya kita bisa langsung membuat pivot table nya
```

```python
monthly_performance = company_revenue_df.groupby('month').agg({
  'order_id':'count', 'payment_value':'sum'
  }).rename(columns={'order_id':'sales', 'payment_value':'revenue'})

print(monthly_performance)
```

```python
#seperti yang terlihat terdapat kenaikan signifikan pada jumlah sales dan revenue yang dimulai dari bulan 2016-09 - 2018-08, meskipun terjadi penurunan tajam di bulan 2018-09 - 2018-10
```

```python
#gabungkan data company_revenue_df dan product_review_df by order_id untuk dijadikan main_df

main_df = pd.merge(
    company_revenue_df, product_review_df, on='order_id'
    )
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Bagaimana demografi pelanggan yang kita miliki?

#untuk menjawab pertanyaan pertama kita perlu membuat pivot table dari dataset customers_df seperti berikut

customer_demographics = customers_df.groupby('customer_state').agg({
    'customer_id':'count'
}).rename(columns={'customer_id':'count'})

# membuat bar chart nya
plt.figure(figsize=(10, 8))
plt.bar(customer_demographics.index, customer_demographics['count'])
plt.xlabel("State")
plt.ylabel("Number of Customers")
plt.title("Customer Demographics by State")
plt.xticks(rotation=45)
plt.show()

```

```python
# dari data diatas kita dapat melihat bahwa kita memiliki konsumen terbanyak yang berasal dari state SP dengan total lebih dari 40.000 customer, sementara RJ diurutan kedua dan MG diurutan ketiga
```

### Pertanyaan 2:

```python
#Bagaimana performa penjualan dan revenue perusahaan dalam beberapa bulan terakhir?

#karena kita sudah membuat pivot table sebelumnya kita bisa langsung membuat visualisasi datanya

plt.figure(figsize=(10, 5))
plt.plot(monthly_performance.index.astype(str), monthly_performance['sales'], label='Sales')
plt.plot(monthly_performance.index.astype(str), monthly_performance['revenue'], label='Revenue')
plt.xlabel("Month")
plt.ylabel("Count")
plt.title("Monthly Sales and Revenue Performance")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#dengan melakukan visualisasi data, kita dapat dengan mudah memahami bawah terjadi progress kenaikan yang sangat bagus dari periode 2016-09 sampai dengan 2018-08, dan dilanjutkan dengan drop yang sangat menghawatirkan pada periode 2018-09 dan 2018-10
```
